```python
# Scraping the Dataset_name from webpage  
try:
    for i in driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a'):      # Extracting Dataset_name from xpath
        Dataset_name.append(i.text)
except NoSuchElementException:
    Dataset_name.append('-')
    
    
# Scraping the Data_type from webpage      
try:
    for i in driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p'):      # Extracting Data_type from xpath
        Data_type.append(i.text)
except NoSuchElementException:
    Data_type.append('-')
    
    
# Scraping the Task from webpage  
try:
    for i in driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p'):      # Extracting Task from xpath
        Task.append(i.text)
except NoSuchElementException:
    Task.append('-')
    
    
# Scraping the Attribute_type from webpage  
try:
    for i in driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p'):      # Extracting Attribute_type from xpath
        Attribute_type.append(i.text)
except NoSuchElementException:
    Attribute_type.append('-')
    
    
    
# Scraping the No_of_instances from webpage      
try:
    for i in driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p'):      # Extracting No_of_instances from xpath
        No_of_instances.append(i.text)
except NoSuchElementException:
    No_of_instances.append('-')
    
    


    
# Scraping the No_of_attribute from webpage      
try:
    for i in driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p'):      # Extracting No_of_attribute from xpath
        No_of_attribute.append(i.text)
except NoSuchElementException:
    No_of_attribute.append('-')
    
    

# Scraping the Year from webpage      
try:
    for i in driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p'):      # Extracting Year from xpath
        Year.append(i.text)
except NoSuchElementException:
    Year.append('-')

    
```

```python
#slicing the column header from the column

Data_type=Data_type[1:]
Task=Task[1:]
Attribute_type=Attribute_type[1:]
No_of_instances=No_of_instances[1:]
No_of_attribute=No_of_attribute[1:]
Year=Year[1:]
```
