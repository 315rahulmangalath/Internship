```python
#creating Lists to add data from webscraping

city_location = []
BRAND = []
price = []
Manuf_yr = []
Insurance_val = []
fuel_type = []
seat_no = []
Kms_driven = []
ownerships = []
Displacement = []
Transmission = []
Mileage = []
```

```python

for i in car_urls.car_urls:              #car_urls scrapped
    driver.get(i)
    time.sleep(2)
    
    #fetching city_location
      
    try:
        city_loc = driver.find_element(By.XPATH,'//input[@id="citySelect"]') 
        city_location.append(city_loc.text)
    except:
        city_location.append("-")
    #fetching bRAND NAME    
    try:
        Name = driver.find_element(By.XPATH,'//div[@class="vehicleName"]') 
        BRAND.append(Name.text[5:])
    except:
        BRAND.append("-")
        
    #fetching price    
    try:
        pri = driver.find_element(By.XPATH,'//div[@class="vehiclePrice"]') 
        price.append(pri.text.split()[1])
    except:
        price.append("-")
    
    
    #fetching Insurance Validity
    try:
        ins_val = driver.find_element(By.XPATH,'//div[@class="label-text"]/div[contains(text(),"Insurance Validity")]//following-sibling::span') 
        Insurance_val.append(ins_val.text)
    except:
        Insurance_val.append("-")
        
        
    #fetching Fuel Type
    try:
        fuel = driver.find_element(By.XPATH,"//div[@class='label-text']/div[contains(text(),'Fuel Type')]//following-sibling::span") 
        fuel_type.append(fuel.text)
    except:
        fuel_type.append("-")
    
    
    #fetching Seats no
    try:
        seat = driver.find_element(By.XPATH,'//div[@class="label-text"]/div[contains(text(),"Seats")]//following-sibling::span') 
        seat_no.append(seat.text.split(' ')[0])
    except:
        seat_no.append("-")
        
     #fetching Kms Driven
    
    
    try:
        Kms = driver.find_element(By.XPATH,'//div[@class="label-text"]/div[contains(text(),"Kms Driven")]//following-sibling::span') 
        Kms_driven.append(Kms.text.split(' ')[0])
    except:
        Kms_driven.append("-")
        
        
     #fetching Ownership
    
    
    try:
        owner = driver.find_element(By.XPATH,'//div[@class="label-text"]/div[contains(text(),"Ownership")]//following-sibling::span') 
        ownerships.append(owner.text)
    except:
        ownerships.append("-")
        
        
        
     #fetching Engine Displacement
    
    try:
        Displ = driver.find_element(By.XPATH,'//div[@class="label-text"]/div[contains(text(),"Engine Displacement")]//following-sibling::span') 
        Displacement.append(Displ.text.split()[0])
    except:
        Displacement.append("-")
        
   
    #fetching Transmission
    
    try:
        transmis= driver.find_element(By.XPATH,'//div[@class="label-text"]/div[contains(text(),"Transmission")]//following-sibling::span') 
        Transmission.append(transmis.text)
    except:
        Transmission.append("-")
    
    
    
        
     #fetching Mileage
    
    try:
        Mileag= driver.find_element(By.XPATH,'//div[@class="label-text"]/div[contains(text(),"Mileage")]//following-sibling::span') 
        Mileage.append(Mileag.text.split()[0])
    except:
        Mileage.append("-")
        
    
    #fetching yr of manufacture
    try:
        man_yr = driver.find_element(By.XPATH,'//div[@class="label-text"]/div[contains(text(),"Year of Manufacture")]//following-sibling::span') 
        Manuf_yr.append(man_yr.text)
    except:
        Manuf_yr.append("-")
```

```python
print(len(city_location),len(BRAND),len(price),len(Manuf_yr),len(fuel_type),len(seat_no),len(Kms_driven),len(ownerships),len(Displacement),len(Transmission),len(Mileage))
```

```python
#Creating Dataframe

Used_Cars=pd.DataFrame({'Car_name':BRAND,'city':city_location,'Manuf_yr':Manuf_yr,'fuel_type':fuel_type,
                            'seat_nos':seat_no,'Kms_driven':Kms_driven,
                            'ownerships':ownerships,'Displacement_in_CC':Displacement,'Transmission':Transmission,
                            'Mileage':Mileage,'price_in_lakhs':price
    
})
```

```python
# Saving the data into excel file
Used_Cars.to_excel("Used_Cars.xlsx")

```
